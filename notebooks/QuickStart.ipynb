```python
%%manim -qm InlineCircle
from manim import *

class InlineCircle(Scene):
    def construct(self):
        c = Circle()
        label = MathTex(r"x^2 + y^2 = r^2")
        label.next_to(c, DOWN)
        self.play(Create(c), Write(label))
        self.wait(1)
```
